```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import pandas as pd
```

```python
model = Sequential()
model.add(Dense(1, input_shape=(2,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
data = pd.read_csv('datos.csv', sep="|")
```

```python
data
```

```python
tuplas = data[['Peso', 'Talla']]
tuplas = tuplas.to_numpy()
tuplas
```

```python
salidas = data[['IMC']]
salidas = salidas.to_numpy()
salidas
```

```python
print("Inicio entrenamiento........:")
history = model.fit(tuplas, salidas, epochs=1000, verbose=True)
print("Fin entrenamiento")
```

```python
result = model.predict([[78, 1.79]])
print(result)
```
